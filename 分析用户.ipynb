```python
import pandas as pd
import os
import glob
import re
from datetime import timedelta
```

```python
csv_files = glob.glob(os.path.join('./datas/users/','*.csv'))
```

```python
dfs = []
```

```python
pattern = r'((?<=./datas/users/new/)([^_]+)(?=_))|(?<=./datas/users/)(.+)(?=\.csv)'
```

```python
mt = re.search(pattern,csv_files[0])
mt.group()
```

```python
for fl in csv_files:
    df = pd.read_csv(fl)
    df['follower'] = re.search(pattern,fl).group()
    dfs.append(df)
```

```python
all_user = pd.concat(dfs)
all_user
```

```python
all_user['collection_date'] = pd.to_datetime(all_user['collection_date'])
all_user =  all_user[all_user['collection_date']> (pd.Timestamp.now() - pd.Timedelta(hours=2))]
```

```python
all_user2 = all_user[['name','username','description','follower','collection_date']]
```

```python
all_user3 = all_user2.sort_values('username').drop_duplicates(subset=['username','follower'])
```

```python
all_user3['follower_count'] = 1
```

```python
all_user4 = all_user3.groupby(['username']).agg({
    'name': 'first',
    'description': 'first',
    'follower': lambda x: ','.join(x),
    'follower_count': 'count'
})
```

```python
all_user4.reset_index(inplace=True)
```

```python
all_user4.sort_values(['follower_count'],ascending=False).head(100)
```
